## meausre the central tendency of the Philly park sizes
measure the mean and the median and compare them

````
meanps = pparks['ACREAGE'].mean()
meanps
````

````
# show the median size of the dataframe
# are they the same?
medianps = pparks['ACREAGE'].median()
medianps
````

**What can you conclude from
 the values of the mean and median you extracted? (discuss in a text cell below)**

From finding the mean and median of the acreage of the parks, we can conclude that the data is very heavily left-skewed. The median size of the parks are only about 3.6 acres, which means most parks are 3.6 arces or less. Since the mean acreage of the parks is so high at about 20 acres, this means that there is a very small number of parks that are very larges sizes. Having a few large numbers in a data set of mostly small numbers can stretch the mean much higher than the median.

hint: to plot a line over the histogram use ```pl.axvline(x=myx)``` and pass to this function
the x value corresponding to the mean and to the median (in my example that would be called myx).
As a default, this will plot a line that spanns the whole y range

````
# visualize a histogram of the park size.
# you can use the method of the ACREAGE column .hist()
# choose an appropriate number of bins with the argument bins=
# overplot on the histogram the mean and the median of the distribution

ax = pparks["ACREAGE"].hist(bins=56)
ax.set_title('Size Distribution of Philly Parks')
ax.set_xlabel('Size of Philly Parks (acres)')
ax.set_ylabel('# of Parks')

ax.axvline(x=meanps, color = 'r', label = 'mean')
ax.axvline(x=medianps, color = 'y', label = 'median')

ax.legend(); #to get the legend to render pass to axvline the argument label="..." with ... being whatever you want displayed
````

Figure 1 shows the distribution of park sizes in Philadelphia. The median of park sizes is 3.6 acres. There are a few outlying parks that are very large and this stretches the mean to 20 acres.

````
# if you need help with this step do ask me about boolean arrays used as indices
# and slicing of arrays and series! remember we have a slack channel #hw2
#acres_pparks = pparks[["ACREAGE"]].copy()
#for x in acres_pparks['ACREAGE']:
  #print(x)
 # if x > 10:
    #print(x)
  #  ax = acres_pparks.droplevel()

pparks = pparks.drop(pparks[pparks['ACREAGE'] > 10].index)

ax = pparks["ACREAGE"].hist(bins=56)
ax.set_title('Size Distribution of Philly Parks < 10 acres')
ax.set_xlabel('Size of Philly Parks (acres)')
ax.set_ylabel('# of Parks')

ax.axvline(x=meanps, color = 'r', label = 'mean')
ax.axvline(x=medianps, color = 'y', label = 'median')

ax.legend();
````

Figure 2 shows the distribution of Philadelphia parks that are under 10 acres in size. The mean size of this sample of parks is 2.95 acres. The median size of the park is 2.17 acres.



the parks listed in each row are in some cases subsets of larger parks. The "parent" park is listed under PARENT_NAME.

The challenge is to visualize the distribution of sizes of the  parent parks. To do that you have to aggregate the parks based on the "PARENT_NAME" value and sum the size of the parks. The methods of the dataframe you want to use are
```.groupby()```
and
```.sum()```

````
ax = pparks.groupby("PARENT_NAME").sum()
ax
#ax['ACREAGE']
````

````
ax = ax["ACREAGE"].hist()

ax.set_title('Size Distribution of Philly Parent Parks < 10 acres')
ax.set_xlabel('Size of Parks (acres)')
ax.set_ylabel('# of Parks')
````

Figure 3 is a table that groups all of the individual parks in Philadelphia (that are under 10 acres) into their parent park. Figure 4 is a histogram that shows the distribution of size of the parent parks.
